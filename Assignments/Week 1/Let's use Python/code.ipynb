```python
THICKNESS = 0.00008
folded_time = 43
```

[Problem 1] Create using the exponentiation arithmetic operator

```python
folded_thickness = THICKNESS * 2**folded_time

print("Thickness: {} meters".format(folded_thickness))
```

[Problem 2] Unit Conversion

```python
# Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} * 10,000 kilometers".format(folded_thickness / 10000000))
print("It is {: .2f} times distance to the Moon".format(folded_thickness / 1000 / 384400))
```

[Problem 3] Create using a for statement

```python
folded_thickness = THICKNESS
for _ in range(0, folded_time):
    folded_thickness = folded_thickness * 2

print("Thickness: {} meters".format(folded_thickness))
```

[Problem 4] Comparison of calculation time

```python
%%timeit
folded_thickness = THICKNESS * 2**folded_time
```

```python
%%timeit
folded_thickness = THICKNESS
for _ in range(0, folded_time):
    folded_thickness = folded_thickness * 2
```

DISCUSSION: The first method runs faster than the second method

[Problem 5] Saving to a list

```python
thickness_value_list = []

folded_thickness = THICKNESS
thickness_value_list.append(folded_thickness)
for _ in range(0, folded_time):
    folded_thickness = folded_thickness * 2
    thickness_value_list.append(folded_thickness)
```

[Problem 6] Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_value_list)
plt.show()
```

DISCUSSION: The thickness increases exponentially respect to number of folds. It started to upsurge dramatically when the number of folds > 35.

[Problem 7] Customizing graphs

```python
'''
VERSION 1: add grid and change font style, size in order to be easier to see and predict values
'''

%matplotlib inline
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("thickness of folded paper", fontdict=font)
plt.xlabel("number of folds", fontdict=font)
plt.ylabel("thickness[m]", fontdict=font)
plt.plot(thickness_value_list)
plt.show()
```

```python
'''
VERSION 2: change tick values font and change thickness from m to km
'''

%matplotlib inline
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("thickness of folded paper", fontdict=font)
plt.xlabel("number of folds", fontdict=font)
plt.ylabel("thickness[km]", fontdict=font)
plt.tick_params(labelsize=12)
thickness_value_list_km = [x / 1000 for x in thickness_value_list]
plt.plot(thickness_value_list_km)
plt.show()
```

```python
'''
VERSION 3: Change line color style
'''

%matplotlib inline
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("thickness of folded paper", fontdict=font)
plt.xlabel("number of folds", fontdict=font)
plt.ylabel("thickness[km]", fontdict=font)
plt.tick_params(labelsize=12)
thickness_value_list_km = [x / 1000 for x in thickness_value_list]
plt.plot(thickness_value_list_km, color= 'green', marker='o', markersize=5)
plt.show()
```
