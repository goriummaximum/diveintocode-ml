```python
import math
import matplotlib.pyplot as plt
```

<h3>[Problem] The day when chestnut buns cover the solar system<h3>

<h4>Assume the chestnut bun is an ideal sphere with radius r with diameter is approximately the breath of adult human hand. Covering an object means the minimum radius of chestnut bun is equal or greater than that object radius. For every 5 minutes, the volume of chestnut buns x2, which infer the new radius is 2**(1/3) greater than the old radius<h4>

```python
HUMAN_HAND_BREATH = 0.0889 #(m)
RADIUS_CHESTNUTS_BUN = HUMAN_HAND_BREATH / 2 #(m)

RADIUS_SOLAR_SYSTEM = 4.545 * 1000000000 * 1000 #(m)
RADIUS_TOKYO_DOME = (46755 / math.pi)**(1 / 2) #(m)

TIME_PERIOD = 5 #(min)
```

```python
def calculate_covered_time(radius_chestnut_bun, radius_object, time_period):
    """
    A function that compute the time needed for the chestnut bun to cover an object.
    --------------
    radius_chestnut_bun: float
        Radius of the chestnut bun (m)
    radius_object: float
        Radius of the object (m)
    time_period: float
        Time (min)
    Returns
    --------------
    time_list: list
        A list holds times the chestnut bun double its volume
    radius_list: list
        A list holds radiu values at a certain time in time list
    """
    radius_list = [radius_chestnut_bun]
    time_list = [0.0]
    while (radius_chestnut_bun < radius_object):
        radius_chestnut_bun = radius_chestnut_bun * 2**(1/3)
        radius_list.append(radius_chestnut_bun)
        time_list.append(time_list[-1] + 5)

    return time_list, radius_list
```

```python
time_list_solar, radius_list_solar = calculate_covered_time(RADIUS_CHESTNUTS_BUN, RADIUS_SOLAR_SYSTEM, TIME_PERIOD)
time_list_dome, radius_list_dome = calculate_covered_time(RADIUS_CHESTNUTS_BUN, RADIUS_TOKYO_DOME, TIME_PERIOD)
```

```python
print("It takes {} minutes to cover the solar system".format(time_list_solar[-1]))
print("It takes {} minutes to cover the Tokyo Dome".format(time_list_dome[-1]))
```

```python
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("Radius of chestnut buns respect to time", fontdict=font)
plt.xlabel("time (min)", fontdict=font)
plt.ylabel("radius (m)", fontdict=font)
plt.tick_params(labelsize=12)
plt.scatter(time_list_solar, radius_list_solar, color= 'red')
plt.hlines(RADIUS_SOLAR_SYSTEM, xmin = time_list_solar[0], xmax = time_list_solar[-1], color="blue", label="Solar system radius")
plt.legend()
plt.show()
```

```python
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("Radius of chestnut buns respect to time", fontdict=font)
plt.xlabel("time (min)", fontdict=font)
plt.ylabel("radius (m)", fontdict=font)
plt.tick_params(labelsize=12)
plt.scatter(time_list_dome, radius_list_dome, color= 'red')
plt.hlines(RADIUS_TOKYO_DOME, xmin = time_list_dome[0], xmax = time_list_dome[-1], color="green", label="Tokyo Dome radius")
plt.legend()
plt.show()
```
