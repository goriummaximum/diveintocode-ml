```python
import matplotlib.pyplot as plt
```

<h3>„ÄêProblem 1] Number of grains of rice on the 100th day<h3>

```python
NUMBER_OF_DAYS = 100
```

```python
day_list = [day for day in range(1, NUMBER_OF_DAYS + 1)]
rice_grains_list = []
total_rice_grains_list = []

for day in day_list:
    rice_grains_list.append(2**(day - 1))
    total_rice_grains_list.append(sum(rice_grains_list))
```

```python
print("Grains of rice received in total on the 100th day: {}".format(total_rice_grains_list[-1]))
```

```python
%matplotlib inline
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("rice grains from day 1 to {}".format(NUMBER_OF_DAYS), fontdict=font)
plt.xlabel("number of days", fontdict=font)
plt.ylabel("number of rice", fontdict=font)
plt.tick_params(labelsize=12)
plt.plot(day_list, rice_grains_list, color= "red", label = "rice received on that day")
plt.plot(day_list, total_rice_grains_list, color= "blue", label = "total rice received by that day")
plt.legend()
plt.show()
```

<h3>„ÄêProblem 2] Number of rice grains outside of the 100th day<h3>

```python
def compute_sorori_shinzaemon(day=100):
    """
    A function that compute the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """

    list_day = [day for day in range(1, day + 1)]
    list_n_grains = []
    list_total_grains = []

    for day in list_day:
        list_n_grains.append(2**(day - 1))
        list_total_grains.append(sum(list_n_grains))
    
    return list_n_grains, list_total_grains
```

```python
specific_day = 10
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day = specific_day)
```

```python
%matplotlib inline
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
plt.grid()
plt.title("rice grains from day 1 to {}".format(specific_day), fontdict=font)
plt.xlabel("number of days", fontdict=font)
plt.ylabel("number of rice", fontdict=font)
plt.tick_params(labelsize=12)
list_day = [day for day in range(1, specific_day + 1)]
plt.plot(list_day, list_n_grains, color= "red", label = "rice received on that day")
plt.plot(list_day, list_total_grains, color= "blue", label = "total rice received by that day")
plt.legend()
plt.show()
```

<h3>„ÄêProblem 3] How many people can live for how many days with the rice we get?<h3>

```python
def compute_living_days(n_rice, n_people, one_rice_weight=0.021, 
                        rice_weight_per_person_per_meal=100.0, meal_per_day=3):
    """
    A function that compute the living day provided the specific amount of rice grains and people.
    Parameteres
    --------------
    n_rice: int
        Number rice grains
    n_people: int
        Number of people
    one_rice_weight: float
        The weight (mg) of one rice grain (default: 0.021)
    rice_weight_per_person_per_meal: float
        The weight (mg) of rice consumed by one people per meal (default: 100.0)
    meal_per_day: int
        Average number of meals that one person takes per day 
    Returns
    --------------
    A truncated integer represent living days that n_people can handle with n_rice.
    """
    return int((n_rice * one_rice_weight) / (n_people * rice_weight_per_person_per_meal * meal_per_day))
```

```python
"""
Let's take the number of grains rice gained from Sorori Shizaemon of 100 days, 66 people are currently in Slack class and some parameters value that I did reseach online.
"""
list_n_grains, list_total_grains = compute_sorori_shinzaemon(100)
n_people = 66
one_rice_weight = 0.021
rice_weight_per_person_per_meal = 100.0
meal_per_day = 3

living_days = compute_living_days(list_total_grains[-1], n_people, one_rice_weight,                                             rice_weight_per_person_per_meal, meal_per_day)
```

```python
print("It takes {} days to consume that amount of rice grains".format(living_days))
```
