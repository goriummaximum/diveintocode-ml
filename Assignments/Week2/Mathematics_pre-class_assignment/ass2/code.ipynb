```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
font = {'family': 'monospace',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
```

<h3>[Problem 1] Creating random numbers<h3>

```python
mean = np.array([-3, 0])
cov_matrix = np.array([[1.0, 0.8], [0.8, 1.0]])

np.random.seed(0)
random_data = np.random.multivariate_normal(mean, cov_matrix, 500)
```

```python
print(random_data.shape)
```

<h3>[Problem 2] Visualization by scatter plot<h3>

```python
plt.grid()
plt.title("Scatter plot of dummy data", fontdict=font)
plt.xlabel("x1", fontdict=font)
plt.ylabel("x2", fontdict=font)
plt.tick_params(labelsize=12)
plt.scatter(random_data[:, 0], random_data[:, 1])
plt.show()
```

<h3>[Problem 3] Visualization by histogram<h3>

```python
plt.grid()
plt.title("Histogram of x1", fontdict=font)
plt.xlabel("x1", fontdict=font)
plt.ylabel("frequency", fontdict=font)
plt.tick_params(labelsize=12)
plt.hist(random_data[:, 0],bins=50, density=True)
plt.xlim(left = -5, right = 3)
plt.show()
```

```python
plt.grid()
plt.title("Histogram of x2", fontdict=font)
plt.xlabel("x2", fontdict=font)
plt.ylabel("frequency", fontdict=font)
plt.tick_params(labelsize=12)
plt.hist(random_data[:, 1],bins=50, density=True, color="red")
plt.xlim(left = -5, right = 3)
plt.show()
```

<h3>[Problem 4] Addition of data<h3>

```python
mean_2 = np.array([0, -3])

np.random.seed(0)
random_data_2 = np.random.multivariate_normal(mean_2, cov_matrix, 500)
```

```python
print(random_data_2.shape)
```

```python
plt.grid()
plt.title("Scatter plot of dummy data", fontdict=font)
plt.xlabel("x1", fontdict=font)
plt.ylabel("x2", fontdict=font)
plt.tick_params(labelsize=12)
plt.scatter(random_data[:, 0], random_data[:, 1], label = "data 1")
plt.scatter(random_data_2[:, 0], random_data_2[:, 1], label="data 2")
plt.legend()
plt.show()
```

<h3>[Problem 5] Data combination<h3>

```python
total_random_data = np.concatenate((random_data, random_data_2), axis=0)
```

```python
print("The shape of the new dataset: {}".format(total_random_data.shape))
print("dataset: \n{}".format(total_random_data))
```

<h3>[Problem 6] Labeling<h3>

```python
label_array = np.concatenate((np.zeros(random_data[:,0].size), np.ones(random_data_2[:,0].size)), axis=0).reshape                                                 (random_data[:,0].size + random_data_2[:,0].size, 1)
total_random_data_new = np.concatenate((total_random_data, label_array), axis = 1)
```

```python
print("The shape of the new dataset: {}".format(total_random_data_new.shape))
print("dataset: \n{}".format(total_random_data_new))
```
